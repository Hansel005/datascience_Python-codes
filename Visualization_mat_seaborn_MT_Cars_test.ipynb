**#Seaborn**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
tips=sns.load_dataset('tips')
```

```
sns.get_dataset_names()
#tips & titanic used for visualization purpose
#Diamonds for homework problem
```

```
tips
```

Univariate --> "One variable" data visualization

```
#Strip plot
#Jitter - To identify or visualize the distribution & the single line
#True=distribution, False=Single line
#Each dot below represnts the tip value in the Y axis
sns.stripplot(y='tip', data=tips, jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```
#Strip plot
#Jitter - To identify or visualize the distribution & the single line
#True=distribution, False=Single line
#Each dot below represnts the tip value in the Y axis
sns.stripplot(y='tip', data=tips, jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.show()
```

```
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```
#Swarm plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.show()

#Each line in the below graph shows how many people gave tips (Example: 2$)
```

```
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip($)')
plt.show()
```

```
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.ylabel('tip($)')
plt.show()
```

**Box and Violin plot**

```
#plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
#Sunday the value of the tip is more in comparison to other days
```

```
#plt.subplot(1, 2, 2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.show()
```

```
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```
plt.subplot(1,3,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```
plt.subplot(3,3,2)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.subplot(3,3,1)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```
#Combining plots
sns.violinplot(x='day', y='tip', data=tips, inner= None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```
#Combining plots
#If we do not mention subplots, by defaults the plots will be combined
sns.violinplot(x='day', y='tip', data=tips, inner= None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter= False)
plt.ylabel('tip($)')
plt.show()
```

**Bivariate --> "Two variables" data visuaization**

```
#Joint plot
#To check if the tip amount is influenced by total bill
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```
#Density plot
#Smaller circle has hgh density
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```
#Pair plot

sns.pairplot(tips)
plt.show()
```

```
sns.pairplot(tips, hue='sex')
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

Loading dataset

```
data=pd.read_csv('/content/train.csv')
data.head()
```

Survival- Survival (0=No; 1=Yes)

class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd) name - Name sex - Sex age - Age sibsp - Number of Siblings/Spouses Aboard parch - Number of Parents/Children Aboard ticket - Ticket Number fare - Passenger Fare cabin - Cabin embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

Data Types

```
print(data.dtypes)
```

Proportion of target (Survived)

```
data.groupby('Survived')['PassengerId'].count()
```

Visual Exploration

```
survived = data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived");
```

```
not_survived = data.loc[data['Survived']==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived");
```

```
sns.pairplot(data.dropna());
```

```
# Pclass vs Survive
grid = sns.FacetGrid(data, col='Survived', row='Pclass', size=2.4, aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```
# Pclass vs Survive
grid = sns.FacetGrid(data, col='Survived', row='Pclass', size=2.4, aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.1, bins=20)
grid.add_legend()
```

```
# Pclass vs Survive
grid = sns.FacetGrid(data, col='Survived', row='Pclass', size=2.4, aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.9, bins=20)
grid.add_legend()
```

```
sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()
plt.title("QueensTown")
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
plt.title("Southampton")
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
#Most of the Passengers embarked from Southampton
```

```
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
plt.savefig("image.png")
```
